```python
t=np.linspace(0,20,100)
t
```

```python
y=np.sin(t)
y
```

```python
plt.plot(t,y)
```

```python
y1 = np.exp(2*t)
```

```python
plt.plot(t,y1)
```

```python
plt.plot(t,y)
plt.plot(t,y1)
```

```python
plt.subplot(1,2,1) #(1,2,1) : 1= row, 2 = colums , 1 = first graph
plt.plot(t,y)
plt.subplot(1,2,2)
plt.plot(t,y1)
```

```python
plt.figure(figsize=(10,8)) # allows you to edit the size of the graphs 
plt.subplot(2,2,1)
plt.plot(t,np.sin(t),marker='x',color='green') # allows u to add markers at the point and colors also 
plt.subplot(2,2,2)
plt.plot(t,np.cos(t))
plt.subplot(2,2,3)
plt.plot(t,np.exp(t))
plt.subplot(2,2,4)
plt.plot(t,np.exp(-t))
```

```python
fig,ax1=plt.subplots() # to plot two axis in one graph 
ax1.plot(t,np.sin(t))
ax2=ax1.twinx()
ax2.plot(t,np.exp(t),color='red')
```
